```python
class Chocolate:
  def __init__(self, cost, additive='', ):
    if cost <= 0:
      raise ValueError('Стоимость должна быть больше 0!')
    self.cost = cost
    self.additive = additive

  def show_my_chocolate(self):
    if self.additive:
      print(f"Шоколадный батончик, добавка: {self.additive}")
    else:
      print("Обычный шоколадный батончик")

  def __str__(self):
    return f'Шоколадный батончик, {self.cost} рублей'
```

```python
choco1 = Chocolate(12, 'клубника')
choco2 = Chocolate(190, 'карамель')
choco3 = Chocolate(34.2)
```
