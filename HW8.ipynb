```python
import requests
import pandas as pd

base_url = "https://www.anapioficeandfire.com/api"
```

#Работа с API


Получить информацию обо всех книгах и сохранить в pd.DataFrame

```python
response_books = requests.get(f"{base_url}/books")
books_data = response_books.json()

books_df = pd.DataFrame(books_data)
books_df
```

Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame

```python
params = {
    'region': 'The Westerlands',
    'hasWords': 'true'
}
response_houses = requests.get(f"{base_url}/houses", params=params)
houses_data = response_houses.json()

houses_df = pd.DataFrame(houses_data)
houses_df
```

Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame

```python
params = {'region': 'The Westerlands'}
response_houses = requests.get(f"{base_url}/houses", params=params)
houses_data = response_houses.json()

houses_df = pd.DataFrame(houses_data)
houses_df
```
