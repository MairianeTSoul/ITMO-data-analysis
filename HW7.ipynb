```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://lifehacker.ru/topics/technology/"
articles = []
```

```python
def get_article_links(page_number):
    url = f"{base_url}?page={page_number}"
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')
    links = []

    for article in soup.find_all('div', class_='article-card__small-wrapper'):
        link_tag = article.find('a', class_='lh-small-article-card__link')
        if link_tag:
            href = link_tag.get('href')
            if href and 'https' not in href:
                links.append(f"https://lifehacker.ru{href}")
    return links
```

```python
def parse_article(url):
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'lxml')

    title_tag = soup.find('h1', class_='article-card__title')
    title = title_tag.text.strip() if title_tag else "Заголовок не найден"

    content_tag = soup.find('div', class_='single-article__post-content')
    content = content_tag.text.strip() if content_tag else "Текст не найден"

    return {"title": title, "content": content, "url": url}


all_articles = []
for page in range(1, 11):
    print(f"Страница {page}")
    article_links = get_article_links(page)

    for link in article_links:
        print(f"Статья: {link}")
        article_data = parse_article(link)
        if article_data:
            all_articles.append(article_data)
```
