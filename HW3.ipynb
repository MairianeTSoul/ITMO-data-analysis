Выведите самое популярное мужское и самое популярное женское имя на корабле

```python
def getFirstName(name):
  if 'Mrs' in name and '(' in name:
    return name.split('(')[1].split()[0]
  else:
    name_part = name.split(",")[1]
    name_words = name_part.split()
    return name_words[1] if len(name_words) > 1 else None
```

```python
df['First_Name'] = df.apply(lambda row: getFirstName(row['Name']), axis=1)
```

```python
most_popular_male_name = df[df['Sex'] == 'male']['First_Name'].mode()[0]
most_popular_female_name = df[df['Sex'] == 'female']['First_Name'].mode()[0]
print("\nСамое популярное мужское имя на корабле:", most_popular_male_name)
print("Самое популярное женское имя на корабле:", most_popular_female_name)
```
